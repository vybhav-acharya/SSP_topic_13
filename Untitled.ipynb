```python

def graph_maker(X,Y,xLabel,yLabel,title):
    import matplotlib.pyplot as plt
    mean_latency=Y
    def best_fit(X, Y):

        xbar = sum(X)/len(X)
        ybar = sum(Y)/len(Y)
        n = len(X) # or len(Y)

        numer = sum([xi*yi for xi,yi in zip(X, Y)]) - n * xbar * ybar
        denum = sum([xi**2 for xi in X]) - n * xbar**2

        b = numer / denum
        a = ybar - b * xbar

        print('best fit line:\ny = {:.2f} + {:.2f}x'.format(a, b))

        return a, b

    # solution
    a, b = best_fit(X, mean_latency)
    #best fit line:
    #y = 0.80 + 0.92x

    # plot points and fit line
    
    plt.scatter(X, mean_latency)
    yfit = [a + b * xi for xi in X]
    plt.plot(X, yfit)
    plt.xlabel(xLabel)
    plt.ylabel(yLabel)
    plt.title(title)

```
